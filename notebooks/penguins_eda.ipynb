```python
from palmerpenguins import load_penguins
import numpy as np
import pandas as pd
import matplotlib.pyplot as plt
import seaborn as sns
```

```python
df = load_penguins()
```

```python
df.head()
```

```python
df.shape
```

```python
df.isna().any().any()
```

```python
df = df.dropna(axis=0)
```

```python
df.shape
```

```python
df.info()
```

```python
df ['species'].value_counts()
```

```python
df['island'].value_counts()
```

```python
# Distribution of bill length

bins = int(np.sqrt(len(df)))
plt.hist(df['bill_length_mm'], bins=bins)
plt.show()
```

```python
# Distribution of bill depth

plt.hist(df['bill_depth_mm'], bins=bins)
plt.show()
```

```python
# Distribution of flipper length

plt.hist(df['flipper_length_mm'], bins=bins)
plt.show()
```

```python
# Distribution of body mass

plt.hist(df['body_mass_g'], bins=bins)
plt.show()
```

```python
sns.kdeplot(data=df, x='body_mass_g', hue='species')
plt.show()
```

```python
# Thinking about the utility of year for predicting species. Does the average body mass change by the year?

df['year'].value_counts()
```

```python
avg_mass = df.groupby(['species','year'])['body_mass_g'].mean().reset_index()
```

```python
sns.barplot(data=avg_mass, x='year', y='body_mass_g', hue='species')
plt.show()
```

```python
# Boxplot

sns.boxplot(data=df, x='flipper_length_mm')
plt.show()
```
